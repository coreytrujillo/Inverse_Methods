```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.integrate import cumtrapz


nu = 9
d = 0.01
x = np.arange(-4,4+d,d)
ft = lambda x: gamma((nu+1)/2)/(gamma(nu/2)*np.sqrt(nu*np.pi))*(1+x**2/nu)**(-(nu+1)/2)

F = ft(x)
G = cumtrapz(F,x,initial=0)
plt.plot(x,F,'b-')
plt.plot(x,G)
plt.show()


print(ft(0))
```
