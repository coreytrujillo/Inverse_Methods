### Problem Statement
*Repeat the following experiment 1000 times: Generate a set of 10 normally distributed random numbers with expected value 10.5 and standard deviation 1. Perform a t-test of the hypothesis $\mu$ = 10 at the 95% confidence level.*

### Deliverables:
a) How many type II errors were committed?

b) What is the approximate power of the t-test with n=10 against the alternative hypothesis $\mu$ = 10.5?

c) Discuss Results

```python
# Import Packages
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
import scipy.stats as st

# Define constants
j = 1001
mu = 10 # "True" mean
tobs = np.zeros((j,1),float)
n = 10
d = n-1 # Degrees of freedom
t0975 = t975 = st.t.ppf(0.975, n-1)
typeII=0

# Loop through j datasets
for i in range(1,j):
    x = rd.standard_normal((10,)) + 10.5
    m = np.mean(x) # Sample mean
    s = np.sqrt(np.sum((x-m)**2)/d) # Sample standard deviation
    plt.plot(x,'.')
    tobs[i,0] = (mu-m)*np.sqrt(n)/s
    if abs(tobs[i,0])>=t0975:
        typeII += 1

# Print type II Errors
print(typeII, 'type II errors were committed')

# Plot Data
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Random Values')
plt.show()

# Plot t-values
a = np.arange(0,j,dtype=float)
b = t0975*np.ones((j,1),float)
plt.plot(abs(tobs),'.')
plt.plot(a,b,'r-')
plt.xlabel('Index')
plt.ylabel('t value')
plt.legend(['|tobs|', 't0.975'])
plt.title('t Values')
plt.show()

# Power of Data
tpow = (1000-typeII)/1000
print('The approximate power of the t-test is: ', tpow)
```

### Discussion

The power of the t-test is reasonable given that given that approximately 2/3 of the observed t values fall below the $t_{0.975}$ threshold. Since the true mean is 10.5 we would expect to get several type II errors.
